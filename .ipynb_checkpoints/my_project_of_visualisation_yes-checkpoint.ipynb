##### Data Cleaning

```python
"""rename some columns"""
df_can.rename(columns={'ObservationDate':'Date','Country/Region':'Country','Province/State':'Province'},inplace=True)
df_can.head()
```

```python
""" Appply data time to column of last Update & Date"""
df_can['Last Update'] = df_can['Last Update'].apply(pd.to_datetime)
df_can['Date'] = df_can['Date'].apply(pd.to_datetime)
""" df_can.drop(['SNo'],axis=1,inplace=True) this function to delete the SNO column , it already deleted """
df_can.head()
```

```python
"""rename rows values"""
df_can.replace('Mainland China', 'China', inplace=True)
df_can.head()
```

```python
df_can.Country
```

```python
df_can.Confirmed
```

```python
""" sorting data by colmun Confirmed mode ascending"""
df_can.sort_values(by=['Confirmed'],ascending = False, inplace=True)
df_can.head()
```

```python
"""Return dataset with country mainland Morocco only """
df_can_Morocco=df_can[df_can['Country']=='Morocco']
df_can_Morocco.head(3)
```

```python
"""recape 2 columns from dataset with sample of 10"""
result=df_can.loc[:,['Country','Confirmed']]  
result2 = result.sample(n=10,replace=True)
result2
```

```python
""" split the date of Last Update column to day and hour """
df_can['Last Update'] = pd.to_datetime(df_can['Last Update'])
df_can['Day'] = df_can['Last Update'].apply(lambda x:x.day)
df_can['month'] = df_can['Last Update'].apply(lambda x:x.month)
df_can[df_can['Day'] == 22]
```

```python
"""get the sum of all columns in day 30 """
df_can[df_can['Day'] == 30].sum()
```

```python
""" Present Gobal sitatuion of all country: confirmed, death and recovered"""
print('Globally Confirmed Cases: ',df_can['Confirmed'].sum())
print('Global Deaths: ',df_can['Deaths'].sum())
print('Globally Recovered Cases: ',df_can['Recovered'].sum())
```

```python
"""COUNTRY WISE CONFIRMED CASES of 2019-nCoV, 2 April 2020"""
countries_cases = df_can.groupby('Country')['Confirmed','Deaths','Recovered'].sum().reset_index()
countries_cases["Mortality Rate (per 100)"] = np.round(100*countries_cases["Deaths"]/countries_cases["Confirmed"],2)
countries_cases.sort_values('Confirmed', ascending= False).style.background_gradient(cmap='YlOrBr',subset=["Confirmed"])\
                        .background_gradient(cmap='Reds',subset=["Deaths"])\
                        .background_gradient(cmap='Greens',subset=["Recovered"])\
                        .background_gradient(cmap='Purples',subset=["Mortality Rate (per 100)"])
```

```python
"""Present the situation of Morocco """
morocco_cases = df_can[df_can['Country']=='Morocco'][["Province","Confirmed","Deaths","Recovered","Date"]]
morocco_cases["Mortality Rate (per 100)"] = np.round(100*morocco_cases["Deaths"]/morocco_cases["Confirmed"],2)
morocco_cases.sort_values('Confirmed', ascending= False).style.background_gradient(cmap='YlOrBr',subset=["Confirmed"])\
                        .background_gradient(cmap='Reds',subset=["Deaths"])\
                        .background_gradient(cmap='Greens',subset=["Recovered"])\
                        .background_gradient(cmap='Purples',subset=["Mortality Rate (per 100)"])
```

# Visualizing Data using Matplotlib<a id="8"></a>

```python
# we are using the inline backend
%matplotlib inline
import seaborn as sns
sns.set_context('paper')
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
print(plt.style.available)
mpl.style.use(['ggplot']) # optional: for ggplot-like style
```

#### let's  how the CoronaVirus is spreading with the days passing by 

```python
data_d = df_can[df_can['Day'] == 22]
#data_d.head()
```

```python
"""Using line plot"""
plt.figure(figsize=(10,6))
plt.xlabel('Days')
plt.ylabel('Confirmed cases')
plt.title('the number of confirmed cases by days')
df_can.groupby('Day').sum()['Confirmed'].plot(kind='line')
```

```python
"""Using barplot"""
plt.figure(figsize=(10,6))
plt.xlabel('Days')
plt.ylabel('Confirmed cases')
plt.title('the number of confirmed cases by days')
sns.barplot(x='Day',y='Confirmed',data=df_can)
```

```python
"""using plot bar , get the confirmed cases in 20 Province defferent"""
plt.figure(figsize=(16,6))
plt.xlabel("Province")
plt.ylabel("Confirmed cases")
plt.title("the confirmed cases in 20 Province")
temp = data_d.groupby(['Province']).sum().head(20)['Confirmed'].plot.bar()
```

```python
df_can.groupby('Province')['Deaths'].sum().reset_index().sort_values(by=['Deaths'],ascending=False).head()
```

```python
"""get the sate of deaths cases by time"""
plt.figure(figsize=(10,6))
plt.title('the number of deaths cases by time')
plt.ylabel('Deaths cases')
df_can.groupby('Date').sum()['Deaths'].plot()
```

```python
"""get the sate of Recovered cases by time"""
plt.figure(figsize=(10,6))
plt.title('the number of deaths Recovered by time')
plt.ylabel('Recovered cases')
df_can.groupby('Date').sum()['Recovered'].plot()
```

### Confrim vs Recovery vs Death histogram


```python
from datetime import date, timedelta
plt.figure(figsize=(23,10))
plt.bar(df_can.Date, df_can.Confirmed,label="Confirm")
plt.xlabel('Date')
plt.ylabel("Count")
plt.legend(frameon=True, fontsize=12)
plt.title('Confrim',fontsize=30)
plt.show()


plt.figure(figsize=(23,10))
plt.bar(df_can.Date, df_can.Recovered,label="Recovery")
plt.xlabel('Date')
plt.ylabel("Count")
plt.legend(frameon=True, fontsize=12)
plt.title('Recovery',fontsize=30)
plt.show()

plt.figure(figsize=(23,10))

plt.bar(df_can.Date, df_can.Deaths,label="Death")
plt.xlabel('Date')
plt.ylabel("Count")
plt.legend(frameon=True, fontsize=12)
plt.title('Death',fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(23,10))
plt.bar(df_can.Date, df_can.Confirmed,label="Confirm")
plt.bar(df_can.Date, df_can.Recovered,label="Recovery")
plt.bar(df_can.Date, df_can.Deaths,label="Death")
plt.xlabel('Date')
plt.ylabel("Count")
plt.legend(frameon=True, fontsize=12)
plt.title('Confrim vs Recovery vs Death',fontsize=30)
plt.show()
```

```python
Total_confirmed = df_can['Confirmed'].sum()
Total_recovered = df_can['Recovered'].sum()
Total_death = df_can['Deaths'].sum()
data = [['Confirmed', Total_confirmed], ['Recovered', Total_recovered], ['Death', Total_death]] 
df = pd.DataFrame(data, columns = ['state', 'count']) 

fig = px.pie(df,
             values="count",
             names="state",
             title="State of all cases around the world",
             template="seaborn")
fig.update_traces(rotation=90, pull=0.05, textinfo="percent+label")
fig.show()
```
